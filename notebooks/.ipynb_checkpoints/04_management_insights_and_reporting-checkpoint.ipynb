```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set_context("talk")
```

```python
use_cols = [
    "SK_ID_CURR",
    "TARGET",
    "AMT_CREDIT",
    "AMT_INCOME_TOTAL",
    "DAYS_BIRTH",
    "EXT_SOURCE_2",
    "CNT_FAM_MEMBERS",
    "REGION_RATING_CLIENT"
]

df = pd.read_csv(
    "../archive/feature_matrix.csv",
    usecols=use_cols,
    nrows=100_000
)

df["AGE_YEARS"] = (-df["DAYS_BIRTH"] / 365).round()

df.shape
```

```python
kpi_snapshot = {
    "Portfolio Size": len(df),
    "Default Rate": df["TARGET"].mean(),
    "Avg Credit": df["AMT_CREDIT"].mean(),
    "Avg Income": df["AMT_INCOME_TOTAL"].mean()
}

pd.DataFrame.from_dict(
    kpi_snapshot,
    orient="index",
    columns=["Value"]
)
```

## Executive KPI Snapshot â€“ Interpretation

These KPIs provide a high-level overview of portfolio health.

Management relevance:
- Default rate signals overall portfolio risk
- Credit and income averages indicate portfolio positioning
- These metrics form the core of monthly management reporting


```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df["AMT_CREDIT"],
    y=df["TARGET"],
    alpha=0.2
)

plt.title("Risk vs Credit Exposure")
plt.xlabel("Credit Amount")
plt.ylabel("Default (0/1)")
plt.show()
```

## Risk vs Exposure Insight

Higher exposure loans carry higher loss potential.

Management takeaway:
- Growth in high-ticket loans must be risk-adjusted
- Portfolio mix matters as much as portfolio size
